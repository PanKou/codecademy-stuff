# U.S. Medical Insurance Costs

```python
import csv
```

```python
# Importing the raw data, I added a 0-indexed column.
raw_data = []
with open('insurance.csv') as in_data:
    reader = csv.DictReader(in_data)
    for row in reader:
        raw_data.append(row)
```

```python
print(raw_data)
```

```python
# Converting the raw sting data into numbers

raw_sexes = []
raw_smoker = []
raw_region = []
con_sexes = []
con_smoker = []
con_regionns = []
con_regionew = []

for person in raw_data:
    raw_sexes.append(person['sex'])
    raw_smoker.append(person['smoker'])
    raw_region.append(person['region'])
for sex in raw_sexes:
    if sex == 'female':
        con_sexes.append(1)
    else:
        con_sexes.append(0)
for smoker in raw_smoker:
    if smoker == 'yes':
        con_smoker.append(1)
    else:
        con_smoker.append(0)
for region in raw_region:
    if region == 'northeast' :
        con_regionns.append(1)
        con_regionew.append(1)
    elif region == 'northwest':
        con_regionns.append(1)
        con_regionew.append(0)
    elif region == 'southeast':
        con_regionns.append(0)
        con_regionew.append(1)
    else:
        con_regionns.append(0)
        con_regionew.append(0)
```

```python
#Creating the cleaned data set.

data = raw_data

for person in raw_data:
    person['index'] = int(person['index'])
    person['age'] = int(person['age'])
    person['bmi'] = float(person['bmi'])
    person['children'] = int(person['children'])
    person['sex'] = con_sexes[int(person['index'])]
    person['smoker'] = con_smoker[int(person['index'])]
    person['regionns'] = con_regionns[int(person['index'])]
    person['regionew'] = con_regionew[int(person['index'])]
    person['charges'] = float(person['charges'])
    person.pop('region')
```

```python
#defining a person class
class Person():
    def __init__(self, age, sex, bmi, children, smoker, regionns, regionew, charges):
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.regionns = regionns
        self.regionew = regionew
        self.charges = charges

# Defining function to calculate the model and then find the |residual|

    def model(self, coeff):
        return self.age*coeff[0] + self.sex*coeff[1] + self.bmi*coeff[2] + self.children*coeff[3] + self.smoker*coeff[4] + self.regionns*coeff[5] + self.regionew*coeff[6] + coeff[7]
    def compare_model(self,coeff):
        return abs(self.charges - self.model(coeff))
            
    def __repr__(self):
        return "{}, {}, {}, {}, {}, {}, {}, {}".format(self.age, self.sex, self.bmi, self.children, self.smoker, self.regionns, self.regionew, self.charges)
```

```python
# Defining a function to find the average

def find_average(key):
    total_key = 0
    for person in data:
        total_key += person[key]
    return total_key/len(data)

# Defining function to find total model error of model. 

def total_model_error(pool, coeff):
    total_error = 0
    for drop in pool:
        total_error += drop.compare_model(coeff)
#     print("Comparing! Total_Error:", total_error)
    return total_error
```

```python
# Defining a function to sweep a range with a set size of diffence in coefficients.
#Sweep steps the step size, range sets the number of steps.

def make_range(initial, sweep, rng):
    return [initial + i * sweep for i in range(-rng, rng+1)]
```

```python
# Making a function to do 1-D linear Regression (more than that is too slow for now). 
# It's obvious from previous testing that the smoker coefficient is the largest one,
# so let's find an initial value of this first. We also need to keep the constant.

def test_model_smoker(pool, initial_coeff, coeff_sweep, rng):
    smokerrange = make_range(initial_coeff[4], coeff_sweep, rng)
    constantrange = make_range(initial_coeff[7], coeff_sweep, rng)
    minimum_error = float('inf')
    min_sm = initial_coeff[4]
    min_const = initial_coeff[7]
    
    for sm in smokerrange:
        for const in constantrange:
            coeff = [initial_coeff[0], initial_coeff[1], initial_coeff[2], initial_coeff[3], sm, initial_coeff[5], initial_coeff[6], const]
            er = total_model_error(pool, coeff)
            if er < minimum_error:
                minimum_error = er
                min_sm = sm
                min_const = const
    return minimum_error, [min_sm, min_const]

def run_smoker(pool, initial_coeff, initial_sweep, rng, n):
    min_error = 0
    min_sm = initial_coeff[4]
    min_const = 0
    sweep = initial_sweep
    coeff = initial_coeff
    for i in range(n):
        coeff = [initial_coeff[0], initial_coeff[1], initial_coeff[2], initial_coeff[3], min_sm, initial_coeff[5], initial_coeff[6], min_const]
#         print("Inserted:", min_sm, min_const)
        min_error, retlist = test_model_smoker(pool, coeff, sweep, rng)
        min_sm = retlist[0]
        min_const = retlist[1]
        sweep = sweep*0.60
#         print("one run done", min_sm, min_const, min_error)
    coeff = [initial_coeff[0], initial_coeff[1], initial_coeff[2], initial_coeff[3], min_sm, initial_coeff[5], initial_coeff[6], min_const]
    return min_sm, min_const, min_error, coeff
```
